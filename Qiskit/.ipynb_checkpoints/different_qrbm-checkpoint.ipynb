```python
from qiskit import Aer, QuantumCircuit, transpile, assemble
from qiskit.visualization import plot_histogram
import numpy as np

# Create a simple RBM circuit
def create_rbm_circuit(num_visible, num_hidden):
    rbm_circuit = QuantumCircuit(num_visible + num_hidden, num_hidden)

    # Apply Hadamard gates on visible units
    rbm_circuit.h(range(num_visible))

    # Apply random rotations on hidden units
    for qubit in range(num_visible, num_visible + num_hidden):
        rbm_circuit.ry(np.random.rand() * 2 * np.pi, qubit)

    # Add CNOT gates for the connections
    for v in range(num_visible):
        for h in range(num_visible, num_visible + num_hidden):
            rbm_circuit.cx(v, h)

    return rbm_circuit

# Define the RBM parameters
num_visible_units = 4
num_hidden_units = 2

# Create RBM circuit
rbm_circuit = create_rbm_circuit(num_visible_units, num_hidden_units)

plot(rbm_circ)

# Transpile the circuit for the simulator
simulator = Aer.get_backend('statevector_simulator')
transpiled_circuit = transpile(rbm_circuit, simulator)

# Run the transpiled circuit on the simulator
result = simulator.run(transpiled_circuit).result()

# Display the final state vector
state_vector = result.get_statevector()
print("Final State Vector:\n", state_vector)

# Plot the histogram of the final state vector
plot_histogram(result.get_counts())
```
