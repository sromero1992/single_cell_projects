```python
from sklearn.datasets import load_iris

iris_data = load_iris()
```

```python
print(iris_data.DESCR)
```

```python
# Obtaining the properties from this dataset
features = iris_data.data
labels = iris_data.target
```

```python
# we'll normalize, applying a simple transformation to represent all features on the same scale.
from sklearn.preprocessing import MinMaxScaler

features = MinMaxScaler().fit_transform(features)
```

```python
# We plot them to see if there is a visual correlation within data points.
import pandas as pd
import seaborn as sns

df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
df["class"] = pd.Series(iris_data.target)

sns.pairplot(df, hue="class", palette="tab10")
```

```python
# Training the classical machine learning model
# split in two parts the dataset for training and testing
from sklearn.model_selection import train_test_split
from qiskit_algorithms.utils import algorithm_globals

algorithm_globals.random_seed = 123
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, train_size=0.8, random_state=algorithm_globals.random_seed
)
```

```python
#We train a classical Support Vector Classifier from scikit-learn
from sklearn.svm import SVC

svc = SVC()
_ = svc.fit(train_features, train_labels)  # suppress printing the return value
```

```python
# Check classical ML model performance
train_score_c4 = svc.score(train_features, train_labels)
test_score_c4 = svc.score(test_features, test_labels)

print(f"Classical SVC on the training dataset: {train_score_c4:.2f}")
print(f"Classical SVC on the test dataset:     {test_score_c4:.2f}")
```

```python
# Now quantum machine learning (QML)
from qiskit.circuit.library import ZZFeatureMap

num_features = features.shape[1]

# Feature map is the encoding of the information
feature_map = ZZFeatureMap(feature_dimension=num_features, reps=1)# we can use other "qiskit.cirbuit.library"
feature_map.decompose().draw(output="mpl", style="clifford", fold=20)
```

```python
from qiskit.circuit.library import RealAmplitudes

# Ansantz is the actual quantum circuit to use, which is real amplitudes here
ansatz = RealAmplitudes(num_qubits=num_features, reps=3) # We can use other "qiskit.cirbuit.library"
ansatz.decompose().draw(output="mpl", style="clifford", fold=20)
```

```python
from qiskit_algorithms.optimizers import COBYLA

# Choosing the optimizing algorithm (gradient free optimizer)
optimizer = COBYLA(maxiter=100)
# We can choose L_BFGS_B but will compute gradients and costs more time
```

```python
from qiskit.primitives import Sampler

# Where to train our classifier (simulator or quantum computer)
# Sampler is statevector based
sampler = Sampler()
```

```python
from matplotlib import pyplot as plt
from IPython.display import clear_output

objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)

# Plotting purposes
def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

```python
# Ready to build the classifier and fit it
# Variational quantum classifier (VQC)
import time
from qiskit_machine_learning.algorithms.classifiers import VQC

vqc = VQC(
    sampler=sampler,
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer,
    callback=callback_graph,
)

# clear objective value history
objective_func_vals = []

start = time.time()
vqc.fit(train_features, train_labels)
elapsed = time.time() - start

print(f"Training time: {round(elapsed)} seconds")
```

```python
train_score_q4 = vqc.score(train_features, train_labels)
test_score_q4 = vqc.score(test_features, test_labels)

print(f"Quantum VQC on the training dataset: {train_score_q4:.2f}")
print(f"Quantum VQC on the test dataset:     {test_score_q4:.2f}")
# We got decent scores
```

```python
# What can we tune to get better models?
# We can tweak parameters, change the reps to repeat more the gate pattern 
# (large reps provide higher number of parameters and entanglement)
# but could overfit the model
```

```python
# If the data contains more features than qbits, we can reduce the number of features
# Reducing the number of features 
# apply PCA transformation from 4 featurs to 2 (educational purposes only...)
from sklearn.decomposition import PCA

features_bak = features
features = PCA(n_components=2).fit_transform(features)

```

```python
plt.rcParams["figure.figsize"] = (6, 6)
sns.scatterplot(x=features[:, 0], y=features[:, 1], hue=labels,  palette="tab10")
```

```python
# Split data to test and train
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, train_size=0.8, random_state=algorithm_globals.random_seed
)

svc.fit(train_features, train_labels)

train_score_c2 = svc.score(train_features, train_labels)
test_score_c2 = svc.score(test_features, test_labels)

print(f"Classical SVC on the training dataset: {train_score_c2:.2f}")
print(f"Classical SVC on the test dataset:     {test_score_c2:.2f}")
```

```python
num_features = features.shape[1]

feature_map = ZZFeatureMap(feature_dimension=num_features, reps=1)
ansatz = RealAmplitudes(num_qubits=num_features, reps=3)
optimizer = COBYLA(maxiter=40)
vqc = VQC(
    sampler=sampler,
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer,
    callback=callback_graph,
)

# clear objective value history
objective_func_vals = []

# make the objective function plot look nicer.
plt.rcParams["figure.figsize"] = (12, 6)


start = time.time()
vqc.fit(train_features, train_labels)
elapsed = time.time() - start

print(f"Training time: {round(elapsed)} seconds")
```

```python
train_score_q2_ra = vqc.score(train_features, train_labels)
test_score_q2_ra = vqc.score(test_features, test_labels)

print(f"Quantum VQC on the training dataset using RealAmplitudes: {train_score_q2_ra:.2f}")
print(f"Quantum VQC on the test dataset using RealAmplitudes:     {test_score_q2_ra:.2f}")
```

```python
#the scores are higher than a fair coin toss but could be better. It cannot be improved given flat tail
# Let's use another ansantz
from qiskit.circuit.library import EfficientSU2

ansatz = EfficientSU2(num_qubits=num_features, reps=3)
optimizer = COBYLA(maxiter=100)

vqc = VQC(
    sampler=sampler,
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer,
    callback=callback_graph,
)

# clear objective value history
objective_func_vals = []

start = time.time()
vqc.fit(train_features, train_labels)
elapsed = time.time() - start

print(f"Training time: {round(elapsed)} seconds")
```

```python
train_score_q2_eff = vqc.score(train_features, train_labels)
test_score_q2_eff = vqc.score(test_features, test_labels)

print(f"Quantum VQC on the training dataset using EfficientSU2: {train_score_q2_eff:.2f}")
print(f"Quantum VQC on the test dataset using EfficientSU2:     {test_score_q2_eff:.2f}")
```

```python
print(f"Model                           | Test Score | Train Score")
print(f"SVC, 4 features                 | {train_score_c4:10.2f} | {test_score_c4:10.2f}")
print(f"VQC, 4 features, RealAmplitudes | {train_score_q4:10.2f} | {test_score_q4:10.2f}")
print(f"----------------------------------------------------------")
print(f"SVC, 2 features                 | {train_score_c2:10.2f} | {test_score_c2:10.2f}")
print(f"VQC, 2 features, RealAmplitudes | {train_score_q2_ra:10.2f} | {test_score_q2_ra:10.2f}")
print(f"VQC, 2 features, EfficientSU2   | {train_score_q2_eff:10.2f} | {test_score_q2_eff:10.2f}")
```
